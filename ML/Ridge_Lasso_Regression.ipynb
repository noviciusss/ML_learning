```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset.csv")
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## missing values
dataset.isnull().sum()
dataset[dataset.isnull().any(axis=1)]
```

the dataset is converted into two sets based on region from 122th index, we can make a
new column based on the Region 

1:"Bejaia Region Dataset"

2:"Sidi-Bel Abbes Region Dataset"



```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
###Removing null values
df=df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Changes the required columns as interger data type

```python
df[['month','day','year','Temperature','RH']]=df[['month','day','year','Temperature','RH']].astype(int)
```

```python
df.info()
```

#### Changing others columns to flot data datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!= 'Classes':
        df[i]=df[i].astype(float)
```
