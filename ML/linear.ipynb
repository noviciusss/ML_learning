```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("priceprediction.csv")
df
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df.Age,df.Salary,color='red',marker='+')
plt.show()
```

```python
age = df.drop(['Salary'],axis=1)
salary =df['Salary']
```

```python
lig = LinearRegression()
lig.fit(age,salary)
```

```python
lig.predict([[27]])
```
