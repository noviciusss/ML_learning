```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
train.groupby('Sex')['Survived'].mean()
```

```python
train.groupby('Embarked')['Survived'].mean()
```

```python
label = {'male':1,'female':0}
train['Sex']=train['Sex'].map(label)
train.head()
```

```python
test['Sex']=test['Sex'].map(label)
test.head()
```

```python
label_e = {'C':0,'Q':1,'S':2}
train['Embarked']=train['Embarked'].map(label_e)
train.head()
```

```python
test['Embarked']=test['Embarked'].map(label_e)
test.head()
```

```python
### Drop not really needed things
train =train.drop(['Name','Ticket','Fare','Cabin'],axis=1)
test =test.drop(['Name','Ticket','Fare','Cabin'],axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['Age'].fillna((train['Age'].mean()),inplace=True)
train.isnull().sum()
```

```python
train['Embarked'].fillna(2,inplace=True)
train.isnull().sum()
```

```python
test['Age'].fillna((test['Age'].mean()),inplace=True)
test.isnull().sum()
```

```python
X_train=train.drop(['Survived'],axis=1)
y_train = train['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classi = DecisionTreeClassifier()
```

```python
tree_classi.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(25,20))
tree.plot_tree(tree_classi,filled =True,fontsize=18)
plt.show()
```

```python
y_predict= tree_classi.predict(test)
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': y_predict})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

### HyperParameter Tunning

```python
param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
gridsearch_dt = GridSearchCV(tree_classi,param_grid=param,cv=5,scoring='accuracy')
```

```python
gridsearch_dt.fit(X_train,y_train)
```

```python
y_predict_cv= gridsearch_dt.predict(test)
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': y_predict_cv})
output.to_csv('submission_cv.csv', index=False)
print("Your submission was successfully saved!")
```

### SVM with hyperParameter

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}
svc = SVC(random_state=42)
gridsearch = GridSearchCV(estimator=svc,param_grid=param_grid,cv=5,scoring='accuracy',n_jobs=-1)
gridsearch.fit(X_train,y_train)
```
