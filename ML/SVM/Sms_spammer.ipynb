```python
try:
    df = pd.read_csv('spam.csv', encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv('spam.csv', encoding='ISO-8859-1') # or 'latin1'
    except UnicodeDecodeError:
        df = pd.read_csv('spam.csv', encoding='Windows-1252')
```

```python
df=df.drop('Unnamed: 4', axis=1)
df=df.drop('Unnamed: 2', axis=1)
```

```python
df.head()
```

```python
df.columns = ['Status','Message']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Status'].value_counts()
```

```python
Status_map = {'ham':0,'spam':1}
df['Status']=df['Status'].map(Status_map)
```

```python
df.head()
```

```python
df['Status'].value_counts()
```

```python
X = df['Message']
y= df['Status']
```

```python
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25,random_state=0)
vector = CountVectorizer()
```

```python
X_train =vector.fit_transform(X_train)
X_test = vector.transform(X_test)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
y_predict = svc.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_predict,y_test))
print(confusion_matrix(y_predict,y_test))
```

```python
from  sklearn.naive_bayes import MultinomialNB
mult = MultinomialNB()
mult.fit(X_train,y_train)
```

```python
y_predict_mul = mult.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_predict_mul,y_test))
print(confusion_matrix(y_predict_mul,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_predict_log = mult.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_predict_log,y_test))
print(confusion_matrix(y_predict_log,y_test))
```

```python
### Check authenticity of dataset
from sklearn.model_selection import cross_val_score
scores = cross_val_score(svc, X_train, y_train, cv=5)
print("CV scores:", scores)
print("Mean:", scores.mean())
```
