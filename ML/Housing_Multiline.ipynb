### Lets prepare dataframe

```python
dataset=pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
dataset['Price']=california.target
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset.corr()
```

```python
import seaborn as sns
sns.heatmap(dataset.corr(),annot=True)
plt.show()
```

## Independent and dependent features --> So what we what as result is dependent

```python
X= dataset.iloc[:,:-1]#independent
y=dataset.iloc[:,-1]# dependent feature
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

## Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
### Slope for coffieciencts
regression.coef_
```

```python
regression.intercept_
```

```python
## Prediction for test data
y_pred= regression.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
### Performance Matrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean abs error: ",mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))      
```

```python
### R squre and adjusted r asqure
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

## What is pickling ?
##### Pickle is a way to serialize and save your trained ML models (or any Python object) to disk  so you can load them back later without retraining.

```python
import pickle
pickle.dump(regression,open('regression.pkl','wb'))
```

```python
model=pickle.load(open('regression.pkl','rb'))
```

```python
model.predict(X_train)
```
